```python
import pandas as pd 
```

```python
details_data = pd.read_excel('./IG_project/practic_details.xlsx')
```

```python
details_data.head(10)
```

```python
ops = list(details_data.columns[1:-1])
ops
```

```python
frequency = list([details_data[x].sum() for x in ops])
frequency
```

```python
hierarchy = [2, 3 ,4 ,4, 2, 5, 6, 3, 7, 4, 5, 4, 6, 3, 3, 4, 6, 7]
```

```python
ops_d = {'Операции': ops}
frequency_d = {'Частота использования': frequency}
hierarchy_d = {'Уровень в иерархии': hierarchy}
```

```python
ops_data = pd.DataFrame(ops_d)
ops_data['Частота использования'] = frequency
ops_data['Уровень в иерархии'] = hierarchy
ops_data.to_excel('./IG_project/op_data.xlsx', index=False)
```

```python
ops_data.head()
```

```python
details_data.head()
```

```python
priority = {}
for detail in details_data['Наименование']:
    op_l = {}
    for operation in details_data.columns[1:-1]:
        complexity, value = details_data.loc[details_data['Наименование']==detail][['Трудоемкость', operation]].values.tolist()[0]
        freq, hier = ops_data.loc[ops_data['Операции']==operation][['Частота использования', 'Уровень в иерархии']].values.tolist()[0]
        op_l[operation] = ((complexity / hier)*freq)*value
    priority[detail] = op_l
    
iuu = pd.DataFrame(priority).T
```

```python
prior_data = pd.read_excel('./IG_project/priorities.xlsx')
```

```python
prior_data.head()
```

```python
recomendations = {}
for operation in prior_data.columns[1:]:
    if prior_data[operation].sum() == 0:
        continue
    det_list = prior_data.sort_values(by=operation)['Наименование'].tolist()[::-1]
    recomendations[operation] = det_list
    
sample = pd.DataFrame(recomendations)
```

```python
sample.to_excel('./IG_project/results.xlsx', index=False)
```
